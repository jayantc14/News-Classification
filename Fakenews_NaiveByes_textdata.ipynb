```
true = pd.read_csv("/content/drive/MyDrive/IIM/True.csv")
true.head()
```

```
fake = pd.read_csv("/content/drive/MyDrive/IIM/Fake.csv")
fake.head()
```

```
true['target'] = 1
true.head()
```

```
fake['target'] = 0
fake.head()
```

```
df = pd.concat([true,fake])
df.head()
```

```
df.reset_index(inplace = True,drop = True)
```

```
df['text']=df['text']+" "+df['title']
df.drop(['subject','date','title'],axis = 1,inplace = True)
```

```
df
```

```
numpy_array = df.a()
X = numpy_array[:,0]
Y = numpy_array[:,1]
Y = Y.astype('int')
print("X")
print(X)
print("Y")
print(Y)
```

```
y = df['target']
X = df.drop(['target'],axis = 1)
```

```
# create an object of CountVectorizer() class 
from sklearn.feature_extraction.text import CountVectorizer 
vec = CountVectorizer( )
```

```
X = df['text']
```

```
X = X.to_numpy()
```

```
vec.fit(X)
vec.vocabulary_
```

```
# removing the stop words
vec = CountVectorizer(stop_words='english' )
vec.fit(X)
vec.vocabulary_
```

```
# printing feature names
print(vec.get_feature_names())
print(len(vec.get_feature_names()))
```

```
# another way of representing the features
X_transformed=vec.transform(X)
X_transformed
```

```
from sklearn.model_selection import train_test_split
```

```
# Splitting the data into train and test
X_train, X_test, y_train, y_test = train_test_split(X_transformed,y, train_size=0.7,test_size=0.3,random_state=100)
```

```
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```
# building a multinomial NB model
from sklearn.naive_bayes import MultinomialNB

# instantiate NB class
mnb=MultinomialNB()

# fitting the model on training data
mnb.fit(X_train,y_train)
```

```
# predicting probabilities of test data
mnb.predict_proba(X_test)
#Predict the response for test dataset
y_pred = mnb.predict(X_test)
```

```
# fit model on training data with default hyperparameters

from sklearn import metrics
# Model Accuracy
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```
print(metrics.confusion_matrix(y_test, y_pred))
```

```
print(metrics.classification_report(y_test, y_pred))
```
